### IJavascript

```bash
npm install -g ijavascript
```
IJavascript 提供了5个可执行程序: `ijsinstall`, `ijsnotebook`, `ijsconsole`, `ijskernel` 和 `ijs`。

* `ijsinstall` registers the IJavascript kernel with Jupyter, so that other tools (e.g. the Jupyter notebook) can invoke it. 
* After running `ijsinstall`, Jupyter notebook users can invoke the Jupyter notebook as usual. `ijsnotebook` is provided for convenience to users of the IPython notebook prior to version 3. ijsnotebook is a wrapper around ipython notebook. 
* `ijsconsole` is provided for convenience to users as a wrapper around ipython console. 
* `ijskernel` is the executable invoked by Jupyter tools (e.g. the notebook) and that appears in the kernel spec that ijsintall creates for IJavascript. You won't need this command, unless you want to create a custom kernel spec.
* `ijs` is provided for backwards-compatibility. It will be removed in the next major-version update. Please, use `ijsinstall` or `ijsnotebook` instead.

## 小提示

1. Run code from a different kernel in a notebooks

如果想要的话，你可以将多个kernel的代码组合到一个 notebook 中。

在每个cell的开头使用相关的魔法命令来声明你想使用的 kernel：


- `%%bash`
- `%%HTML`
- `%%python2`
- `%%python3`
- `%%ruby`
- `%%perl`


1. Running R and Python in the same notebook

对于此最好的解决方案是安装 rpy2， 通过pip可以很容易安装：

```bash
pip install rpy2
```

可以同时使用这两种语言，甚至在它们之间传递变量：

一个很好的示例 [Revolutions Blog](http://blog.revolutionanalytics.com/2016/01/pipelining-r-python.html)

2. Create a presentation from a Jupyter notebook

Damian Avila 的 RISE 允许你从已有的一个 notebook 创建一个 powerpoint 风格的报告。

你可以通过conda安装RISE:
```bash
conda install -c damianavila82 rise
```
或通过pip:
```bash
pip install RISE
```

然后执行下面的代码安装并启用扩展：
```bash
jupyter-nbextension install rise --py --sys-prefix
jupyter-nbextension enable rise --py --sys-prefix
```

3. 共享 notebook

分享notebook最简单的方式是直接使用notebook文件(.ipynb).不过对于那些不使用Jupyter的人来说，也可以这么做：

* 将notebook保存到比如dropbox中，然后将连接放到 [nbviewer](http://blog.csdn.net/simple_the_best/article/details/nbviewer.jupyter.org). nbviewer将会渲染你存储在任何地方的notebook.
* 安装 [jupyterhub](https://github.com/jupyterhub/jupyterhub), 当你组织一个小型课程或研讨会没有精力关心学生的机器状况时，这非常方便。
